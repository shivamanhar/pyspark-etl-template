```python
%sql
DROP TABLE IF EXISTS silver_fact_sales;
CREATE TABLE IF NOT EXISTS silver_fact_sales (
  sales_id INT,
  transaction_ts STRING,
  date_key  INT,
  product_key INT,
  customer_key  INT,
  store_key INT,
  payment_key  INT,
  quantity INT,
  unit_price DECIMAL(10,2),
  discount_pct DECIMAL(5,2),
  net_sales_amount DECIMAL(10,2),
  tax_amount DECIMAL(10,2)  
)
```
