```python
from pyspark.dbutils import DBUtils
dbutils = DBUtils(spark)
```

```python
source_path = "/Volumes/workspace/etl_practice/my_file/sales_source/"
folders = {
    ",": "/Volumes/workspace/etl_practice/my_file/sales/comma/",
    ";": "/Volumes/workspace/etl_practice/my_file/sales/semicolon/",
    "|": "/Volumes/workspace/etl_practice/my_file/sales/pipe/"
}
```

```python

files = dbutils.fs.ls(source_path)

for f in files:
    if f.name.endswith(".csv"):
        first_line = dbutils.fs.head(f.path, 1024)
        delim = None
        for d in folders.keys():
            if d in first_line:
                delim = d
                break
        if delim:
            dbutils.fs.mv(f.path, folders[delim] + f.name)
```
